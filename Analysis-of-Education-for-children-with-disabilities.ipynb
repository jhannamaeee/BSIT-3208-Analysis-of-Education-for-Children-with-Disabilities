```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set()
```

```python
dt = pd.read_csv (r'C:\Users\arvin\Documents\Python Scripts\Completion-Primary.csv')
dt.head()
```

```python
#Percentage by Development Regions

count_test = dt['Development regions'].value_counts()
label = dt['Development regions'].value_counts().index
plt.figure(figsize=(20,10))
plt.pie(count_test, labels=label, autopct='%1.1f%%')
plt.legend(label)
plt.title("Percentage by Development Regions", fontweight="bold")
plt.show()
```

```python
sns.scatterplot(data=dt, x="Development regions", y="Children with functional difficulties", hue="Category" )
plt.title("Children with functional difficulties in Completion Primary", fontweight="bold")
```

```python
sns.scatterplot(data=dt, x="Development regions", y="Children without functional difficulties", hue="Category" )
plt.title("Children without functional difficulties in Completion Primary", fontweight="bold")
```
